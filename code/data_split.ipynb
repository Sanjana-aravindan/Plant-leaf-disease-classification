```python
import os
import numpy as np
import pandas as pd
from shutil import copyfile
from tqdm import tqdm

INPUT_SIZE = 256
SET = [0.6] #, 0.5, 0.3]

for TRAIN_SIZE in SET:

    # Set input directory
    data_dir = "../segment/"

    # Get classes
    classes = os.listdir(data_dir)
    num_classes = len(classes)
    print(f"Number of classes: {num_classes}")

    # Create train/validation directories
    name = str(int(TRAIN_SIZE * 100)) + '_' + str(int((1-TRAIN_SIZE-0.2) * 100)) + '_'+str(int((1-TRAIN_SIZE-0.2) * 100))+"segmented"
    print (name)
    os.mkdir(data_dir + name)
    train_dir = data_dir + name + '/train/'
    val_dir = data_dir + name + '/validation/'
    test_dir = data_dir + name + '/test/'
    valid_dir = data_dir + name + '/validate/'
    os.mkdir(train_dir)
    os.mkdir(val_dir)
    os.mkdir(test_dir)
    os.mkdir(valid_dir)    
```

```python
for c in classes:
    os.mkdir(train_dir + c)
    os.mkdir(val_dir + c)
    os.mkdir(test_dir+c)
    os.mkdir(valid_dir+c)
```

```python
#     # Get path and label for each image
db = []
index_list=[]
for label, class_name in enumerate(classes):
    path = data_dir + '' + class_name
    for file in os.listdir(path):
        if '.ini' not in file:
            db.append(['{}/{}'.format(class_name, file), label, class_name])
            index_list.append(class_name+"/"+file)
db = pd.DataFrame(db, columns=['file', 'label', 'class_name'], index=index_list)
num_images = len(db)
print(f"Number of images: {num_images}")
```

```python
np.random.seed(87)
msk = np.random.binomial(1, TRAIN_SIZE, num_images)
print (sum(msk))
```

```python
#     # Import images
i = 0
db_val=[]
for file in tqdm(db['file'].values):
    print (file)
    from_path = data_dir + file
    if msk[i] == 1:
        to_path = train_dir + file
    else:
        db_val.append([db.loc[file,'file'], db.loc[file,'label'],db.loc[file,'class_name']])
        to_path = val_dir + file

    copyfile(from_path, to_path)
    i += 1

        
```
